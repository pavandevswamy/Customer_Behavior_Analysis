```python
df["Review Rating"].fillna(df["Review Rating"].mean(),inplace=True)
```

```python
df["Review Rating"].isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df.rename(columns={'purchase_amount_(usd)':'purchase_amount'},inplace=True)
df.columns
```

```python
# create a column age_group

labels=["young adult","adult","middle aged","seniour"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)            # qcut is made equally groups
```

```python
df[["age_group","age"]].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping={
    'Weekly':7,
    "Fortnightly":14,
    "Bi-Weekly":14,
    "Monthly":30,
    "Quarterly":90,
    "Annually":365,
    "Every 3 Months":90
}

df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]]
```
